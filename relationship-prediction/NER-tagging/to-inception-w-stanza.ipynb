```python
import os
from cassis import *
```

```python
import stanza
stanza.download('en')       # This downloads the English models for the neural pipeline

nlp = stanza.Pipeline('en') # This sets up a default neural pipeline in English
```

```python
path = './to-tag'

# create folder for output
os.mkdir('tagged')


for filename in os.listdir(path):
  # hidden file check
  if not filename.startswith('.'):
    full_path = path + "/" + filename
    with open(full_path, 'r') as file:
        text = file.read()

    # create stanza obj
    doc = nlp(text)

    # create CAS
    SENTENCE_TYPE = "de.tudarmstadt.ukp.dkpro.core.api.segmentation.type.Sentence"
    TOKEN_TYPE = "de.tudarmstadt.ukp.dkpro.core.api.segmentation.type.Token"
    NER_TYPE = "webanno.custom.StanzaNER" # internal name in layer settings 

    # load type system
    with open("typesystem.xml", "rb") as f:
      ts = load_typesystem(f)

    cas = Cas(typesystem=ts)
    cas.sofa_string = text

    Sentence = ts.get_type(SENTENCE_TYPE)
    Token = ts.get_type(TOKEN_TYPE)
    NamedEntity = ts.get_type(NER_TYPE)

    # get full sentences from obj
    for sentence in doc.sentences:
      start_char = 0
      end_char = 0
      
      s = sentence.to_dict()
      last_token = len(s)
      
      for d in s:
        if d["id"] == 1:
          start_char = d["start_char"]
        elif d["id"] == last_token:
          end_char = d["end_char"]

      cas_sentence = Sentence(begin=start_char, end=end_char)
      cas.add(cas_sentence)
      print("FULL SENTENCE: " + sentence.text)
      assert sentence.text == cas_sentence.get_covered_text()

    # get all tokens
    all_tokens = []
    for sentence in doc.sentences:
      sentence = sentence.to_dict()
      for d in sentence:
        all_tokens.append(d)

    for token in all_tokens:
      cas_token = Token(begin=token["start_char"], end=token["end_char"])
      cas.add(cas_token)
      print("TOKEN: " + token["text"])
      assert token["text"] == cas_token.get_covered_text()

    # get entities
    for named_entity in doc.ents:
        cas_named_entity = NamedEntity(begin=named_entity.start_char, end=named_entity.end_char, value = named_entity.type)
        cas.add(cas_named_entity)
        print("ENTITY: " + named_entity.text)
        assert named_entity.text == cas_named_entity.get_covered_text()

    expt_name = "./tagged/" + os.path.splitext(filename)[0] + "-anno.xmi"
    cas.to_xmi(expt_name)
```
